```python
import requests
import json
import time
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

map = Basemap()

map.drawcoastlines()
map.drawcountries()
map.drawstates()
map.fillcontinents(color='lightgray',zorder=0)
map.drawmapboundary(fill_color='white')

lons = []
lats = []

start_time = time.time()
end_time = start_time + 3600  # 1 hour in seconds

while time.time() < end_time:
    # fetch data from the producer
    response = requests.get('http://api.open-notify.org/iss-now.json')
    data = json.loads(response.text)

    # extract longitude and latitude from the data
    lon = float(data['iss_position']['longitude'])
    lat = float(data['iss_position']['latitude'])
    
    lons.append(lon)
    lats.append(lat)
    
    # plot the satellite location on the map
    x, y = map(lon, lat)
    map.plot(x, y, marker='o', markersize=1, color='red')
    
    time.sleep(5)

plt.title('Satellite Tracking for 1 Hour')
plt.show()
```
